```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for the uniform distribution
a = -1
b = 1

# Generate data for the uniform distribution
uniform_data = np.random.uniform(a, b, 10000)

# Generate data for the standard normal distribution
normal_data = np.random.randn(10000)

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

# Plot the uniform distribution
axs[0].hist(uniform_data, bins=30, density=True, color='skyblue', edgecolor='black')
axs[0].set_title('Uniform Distribution')

# Plot the standard normal distribution
axs[1].hist(normal_data, bins=30, density=True, color='salmon', edgecolor='black')
axs[1].set_title('Standard Normal Distribution')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Define the time variable t
# t = np.linspace(0, 1, 1000)  # Generate 1000 points between 0 and 1
t = np.arange(0.0, 5.0, 0.01)

# Compute the sine wave
s = np.sin(2 * np.pi * t)

# Plot the sine wave and get the line object
line, = ax.plot(t, s, lw=2)

# Customize the plot
ax.set_xlabel('Time (t)')
ax.set_ylabel('Amplitude')
ax.set_title('Sine Wave')
ax.grid(True)
ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Show the plot
plt.show()
```
