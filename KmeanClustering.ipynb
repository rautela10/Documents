```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

# Generate sample data
X, Y = make_classification(n_samples=100, n_features=1, n_informative=1, n_redundant=0, n_clusters_per_class=1, random_state=42)

# Create a logistic regression model
model = LogisticRegression()

# Train the model
model.fit(X, Y)

# Make predictions
Y_pred = model.predict(X)

# Plot the data points and the decision boundary
plt.scatter(X, Y, color='blue', label='Actual')
plt.scatter(X, Y_pred, color='red', marker='x', label='Predicted')
plt.title('Logistic Regression Classification')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Create a K-Means clustering model with 3 clusters
model = KMeans(n_clusters=3)

# Train the model
model.fit(X)

# Get cluster assignments for each data point
labels = model.labels_

# Plot the data points colored by their cluster assignments
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
